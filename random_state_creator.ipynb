```python
import random
import numpy as np
import pandas as pd
import math
```

```python
# общее число игроков
n=24
# желаемое число команд
t=4
# количество раундов
r=5
# получаемое число игроков в команде 
k=-math.floor(-n/t)

# штраф за пересечения
m=1

# список для распредления игроков
players_to_teams_lst=[]
for tpl in range(t):
    for kpl in range(k):
        players_to_teams_lst.append(tpl)
```

```python
players_to_teams_lst
```

```python
# класс со всеми параметрами игрока
class player():  
    def __init__(self, id=0, score=0.0, teammates=[]):
        self.id= id
        self.score= score
        self.teammates=teammates
        
# создаём игроков и пару нужных массивов
main_player_list=[]
ids_list=[]
scs_list=[]
tmms_list=[]
for i in range(n):
    main_player_list.append(player(id=i,  score=0.0))
    # id
    ids_list.append(main_player_list[i].id)
    # score
    scs_list.append(main_player_list[i].score)
    # лист сокомандников
    tmms_list.append(main_player_list[i].teammates)
```

```python
def penalty_count(new_tms, old_tms, m):
    pen=0
    for nt in new_tms:
        if nt in old_tms:
            pen=pen+m
    return pen
```

```python
def make_sample(ids_list, main_player_list):
    players_list=np.array(ids_list)
    random.shuffle(players_list)

    sample=np.array(np.transpose([players_list, players_to_teams_lst]))

    sample=sample[sample[:,0].argsort()]
    
    teammates=[]

    penalty=0
    for pair in sample:
        player_id=pair[0]
        cur_team=pair[1]
        cur_teammates_lst=[]
        last_teammates=main_player_list[player_id].teammates
        for pair_opp in sample:
            if (pair_opp[1]==cur_team) and (pair_opp[0]!=player_id):
                cur_teammates_lst.append(pair_opp[0])
        teammates.append(cur_teammates_lst)
        penalty=penalty+penalty_count(cur_teammates_lst, last_teammates, m)
        
    return sample, penalty, teammates
```

```python
def make_round(main_player_list, sample, round_number):
    tbl_round=pd.DataFrame(sample[0], columns=['player_id', 'team_id'])
    tbl_round['round_number']=round_number

    for i in tbl_round['player_id'].values: 
        for pl in sample[2][i]:
            main_player_list[i].teammates=np.append(main_player_list[i].teammates, pl)
    return tbl_round
```

```python
tbl=pd.DataFrame()

for rnd in range(r):
    array_list=[]
    penalty_list=[]
    teammates_list=[]
    for tr in range(100000):
        sample=make_sample(ids_list, main_player_list)
        array_list.append(sample[0])
        penalty_list.append(sample[1])
        teammates_list.append(sample[2])
        if sample[1]==0:
            print('ready!')
            break
    le=np.argmin(penalty_list)
    print(penalty_list[le])
    sample_to_work=array_list[le], penalty_list[le], teammates_list[le]

    tbl_round=make_round(main_player_list, sample_to_work, rnd)

    tbl=pd.concat([tbl, tbl_round])
```
